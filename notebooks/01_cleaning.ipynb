```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/raw/Sample - Superstore.csv', encoding='windows-1252')

print("Shape: ", df.shape)          # how many rows and columns
print("DTypes: ",df.dtypes)         # column data types
print("Head: ",df.head())         # first 5 rows
print("Is Null? ",df.isnull().sum()) # missing values per column
```

```python
print(df.isnull().sum())
```

```python


# Check which format your file uses by printing 
print(df['Order Date'].head())
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%Y')
df['Ship Date']  = pd.to_datetime(df['Ship Date'],  format='%m/%d/%Y')

# Verify it worked
print(df[['Order Date', 'Ship Date']].dtypes)
print(df[['Order Date', 'Ship Date']].head())
```

```python
# Time features
df['order_year']    = df['Order Date'].dt.year
df['order_month']   = df['Order Date'].dt.month
df['order_quarter'] = df['Order Date'].dt.quarter
df['days_to_ship']  = (df['Ship Date'] - df['Order Date']).dt.days

# Financial features
df['profit_margin_pct'] = (df['Profit'] / df['Sales'] * 100).round(2)
df['is_loss']           = df['Profit'] < 0
df['discounted']        = df['Discount'] > 0

print(df[['Order Date', 'Ship Date', 'days_to_ship', 
          'profit_margin_pct', 'is_loss', 'discounted']].head(10))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Sales by category
cat_sales = df.groupby('Category')['Sales'].sum().sort_values()
axes[0,0].barh(cat_sales.index, cat_sales.values, color=['#378ADD','#1D9E75','#D85A30'])
axes[0,0].set_title('Total Sales by Category')
axes[0,0].set_xlabel('Sales ($)')

# Monthly sales trend
monthly = df.groupby(['order_year','order_month'])['Sales'].sum().reset_index()
monthly['period'] = monthly['order_year'].astype(str) + '-' + monthly['order_month'].astype(str).str.zfill(2)
axes[0,1].plot(monthly['period'], monthly['Sales'], color='#378ADD', linewidth=1.5)
axes[0,1].set_title('Monthly Sales Trend')
axes[0,1].tick_params(axis='x', rotation=90, labelsize=7)

# Discount vs Profit
axes[1,0].scatter(df['Discount'], df['Profit'], alpha=0.3, color='#D85A30', s=10)
axes[1,0].axhline(0, color='black', linewidth=0.8, linestyle='--')
axes[1,0].set_title('Discount vs Profit')
axes[1,0].set_xlabel('Discount')
axes[1,0].set_ylabel('Profit ($)')

# Profit margin by sub-category
sub_margin = df.groupby('Sub-Category')['profit_margin_pct'].mean().sort_values()
colors = ['#E24B4A' if x < 0 else '#1D9E75' for x in sub_margin.values]
axes[1,1].barh(sub_margin.index, sub_margin.values, color=colors)
axes[1,1].axvline(0, color='black', linewidth=0.8)
axes[1,1].set_title('Avg Profit Margin % by Sub-Category')
axes[1,1].set_xlabel('Margin %')

plt.tight_layout()
plt.show()
```
